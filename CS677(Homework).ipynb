6. On increasing the number of hidden layers to four with each hidden layer having 8 neurons we notice that the overall perfomance of the model decreases.
Moreover we might have a problem of vanishing gradient. Since backpropagation is used to find neural network gradients, the derivatives of each layer are multiplied down the network by the chain rule to compute the derivatives of the original layers. As a result, as we descend to the original levels, the gradient diminishes exponentially. 
As a result we can clearly see that the neurons in the top layers (on the right) evolve faster than neurons in the lower layers (on the left).
