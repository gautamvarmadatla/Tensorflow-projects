### What is Machine Learning?

In the simplest description, it is a science of 

* creating a program to learn from existing data 


* be able to predict a required output when fed by a new data. 


* the driving force behind the widespread application of Artificial Intelligence(AI) as we know today.

### Artificial Neural Network

* There are a lot of algorithms for machine learning. 


* A popular one is Artificial Neural Network. 




#### *input layer* 

consists of two neurons denoted by the green circles. The number of neurons must be equal to the number of features of the dataset being used.

#### *hidden layer* 

consists of the five cyan circles, is arbitrary. These arbitrary parameters are commonly called hyperparameters. Determining the number of layers and the number of neurons is the developer's decision. This is how the term *training* in machine learning came up. We have to *train* over and over our *machine* with different number of hidden layers and neurons to *learn* from our input data. Try 1 hidden layer with 64 neurons first. Then start adding or removing neurons in $n^2$. Thus, you can try 8, 16, 32, 64, 128, 256 and so on. 10, 50, 100 neurons can be used perfectly. But because binary system has been used since the dawn of computers, it seems logical to choose the $n^2$ values. If you decide to use more than one hidden layer, it is advised to use the same number of neurons for each hidden layer. Usually, adding more than one hidden layer is better. 

#### *output layer*

denoted by the yellow circle, must be equal to the number of predicted label or result. Example, if we are trying to predict the (1)price of a house, the number of neurons would be 1. Likewise, if our neural network is trying to classify if a review is (1)good or (2)bad, our neurons would be 2. If, on the other hand, we are classifying a news if it is (1)political, (2)health, (3)entertainment or (4)sports news, then our neurons would be 4.

#### Let's deep dive into the neural network

* This is the *brain* of machine learning. It is inspired by how the "neurons" in our brain works. 


* Each Neural Network consists of the *Input Layer*, *Hidden Layer* and an *Output Layer*. During our learning/training process, this is commonly referred to as model. But technically the model is the result of the process.



**Loss**

This the summation of errors during training and validation. It is safe to say that the lower the loss, the better our model.

**Bias**

This is an extra neuron with a value of 1 added to the input of a layer in an artificial neural network. This is added so that it can be able to *train* our model better. It is best discussed technically [here](https://stackoverflow.com/questions/2480650/role-of-bias-in-neural-networks).

* Mathematically, a bias value allows you to shift the activation function to the left or right, which may be critical for successful learning.




* The output of the network is computed by multiplying the input (x) by the weight (w0) and passing the result through some kind of activation function (e.g. a sigmoid function.)

* Just changing the steepness of the sigmoid won't really work -- you want to be able to shift the entire curve to the right.


* That's exactly what the bias allows you to do. If we add a bias to that network, like so:

**Supervised vs Unsupervised Learning**

* Supervised learning means that we have to "teach" our machine to learn something. We have to provide "training data" which contains raw data and ground truth answers.


* Unsupervised learning means that the machine will "learn" by itself. Data doesn't contain answers.


* An easy way to determine if we will using supervised or unsupervised learning is to check the dataset given. 
    * If the dataset is labeled, like in the example of house price prediction, the prices of the house are given and we are trying to predict the price for new house, it is a supervised learning. 
    * If the dataset is not labeled and we are deriving some relationships in our dataset, this is unsupervised learning. An example is a dataset of customer purchases and we would like to do customer segmentation based on purchased products.


* There are other kinds of learning like reinforcement learning (later in the course), but today we will focus on these two basic learning types.

### Step 1. Import all the modules we need

We need some supporting modules in order to process the data, displaying visual aids and other functions. If the module is not available in your python environment, you can install it via pip.

```python
import keras #the keras high-level API
import pandas as pd #for manipulating dataset
import seaborn as sns #plotting data
import matplotlib.pyplot as plt #plotting data
import sklearn.metrics as metrics #for computing metrics after the training
from sklearn.model_selection import train_test_split #function to split our data
from keras.models import Sequential #neural network model
from keras.layers import Dense #neural network layers
# from keras.optimizers import RMSprop #optimizer for linear regression if you use old version
from tensorflow.keras.optimizers import RMSprop
from keras.callbacks import ModelCheckpoint, EarlyStopping #callbacks during training
```

### Step 2. Read the data into dataframe

* read the data from file using pandas


* first let's define the name of the columns as described in auto-mpg.names


* Always display the top 5 rows to check if we are reading it correctly

### Step 9. Split data into training and testing datasets.

* A good practice is to train our model with a training data and test separately with a test data. 


* We should **never** use the training data to test our model. 


* We should **never** use the testing data to train our model. This can be very tricky. 


* If we test our model with the training data, obviously, it gives us a good result. But what we are after is how well it will predict unseen data which is our test data that was never used in the training.




**Good Practice**

* shuffle the dataset


* so that we do not end up training to a just speficic set of data specially if the dataset is sorted


* the function automatically randomize the dataset


* but we have to set the random_state to a fixed integer 


* so that we can reproduce with the same arrangement of dataset every time


* test_size means that we will create a 20% test data which is a good ratio to start with

```python
df_train, df_test = train_test_split(df_encoded, test_size=0.2, random_state=0)

#get the training label
y_train = df_train["mpg"]

#and the test label
y_test = df_test["mpg"]

#remove the label from train and test data
df_train.pop("mpg")
df_test.pop("mpg")

#display top 5
df_train.head()
#you may have noticed that the index, the left most column is already shuffled
```

### Step 10. Normalize the data

* Normalization is the process of transforming all the values of our features to as close as possible. We can train our model without normalization but learning may take a very long time or if we are in bad luck, our model may not converge at all.


* A good rule of thumb is if difference between features is in hundreds, it is better to normalize the data. Especially for big dataset, we should normalize all features.


* It should be noted that the training data and test data should be normalized with the same technique <b>on the same scale. See the code below that uses the scale from training data to normalize both training and testing data.</b>


* There are many techniques for normalization. One of my favorites is by subtracting the minimum value then dividing it with the maximum value.

```python
#before normalization let's get the statistics of the dataset to determine the min/max value of each column
df_stat = df_train.describe()

#let's transpose the statistics.
#that is writing rows(count, mean, max, etc) as columns and vice versa
df_stat = df_stat.transpose()

#display the statistics
df_stat
```

* <font color='red'>We use the same df_stat to normalize both training and testing data</font>


* define a function to do the normalization

```python
def normalize(dataset, stats):
    '''Returns a normalized dataframe.'''
    return (dataset - stats["mean"])/ stats["max"]
```

* normalize the training and test data to temporary objects


* do not mess up the original dataset


* especially when the data preprocessing above is slow. You don't want to run everything from the beginning again. 

```python
x_train_temp_norm = normalize(df_train, df_stat)
x_test_temp_norm = normalize(df_test, df_stat)

#let's make a copy of our temporary train and test data
x_train =  x_train_temp_norm.copy()
x_test = x_test_temp_norm.copy()

#display top 5 rows
x_train.head(5)
```

### Step 11. Finally, we can create a model now. We are almost done.

Now that we are finished processing our dataset. We can start machine learning.

* create the model of neural network for a regression problem


* keras Sequential model is a linear stack of layers and it is the model that works most of the time


* you must start by creating a model first before stacking it with layers

```python
model = Sequential()
```

* add input layer


* Dense is just a regular densely-connected neural layer


* units = the number of neurons


* activation = defines an output signal that can be fed to the input signal of the next layer


* it "connects" one layer to another


* relu = Rectified Learning Unit


* there are other types of activation but this is the most widely used for hidden layers


* relu cannot be used in the output layer


* adding the first layer actually adds a hidden layer with neurons equal to number specified in units, in this case 64


* the input_dim parameter creates the input layer with neurons, this is the number of features of our training data

```python
INPUT_DIM = x_train.shape[1]
model.add(Dense(units=64, input_dim=INPUT_DIM, activation="relu"))
```

* add another hidden layer with the same neurons as the previous one

```python
model.add(Dense(units=64,  activation="relu"))
```

* Add output layer


* There is only 1 unit here because our output is only one, in this project it is the MPG


* Activation is set to linear


* For problems with predicting a single real value (regression), a linear activation is used. It just outputs whatever the values from the previous layer lineary


* We no longer specify the input_dim because it is implicit, and keras will automatically assign it based from previous layer

```python
model.add(Dense(units=1, activation="linear"))
```

* after adding all the layers, let's compile the model.

```python
model.compile(
              #specify the type of loss to check, that is Mean Squared Error
              loss="mse",          
              #optimization algorithm to use to find the minimum loss/convergence
              optimizer= RMSprop(),
              #what metrics to use to test our validation data
              #Mean Absolute Value
              #Instead of evaluating the squared error, we'll use the absolute value
              #to be able to check the difference between the actual and our prediction
              metrics=["mae"])

#display the summary of our model
model.summary()
```

Let's understand our model summary.

* The first layer shows a total Param of 640. This is the product of the numer of input neurons multipled by the output neurons.


* In our model, we defined the input layer as having 9 input neurons(cylinders, displacement, horsepower, weight, acceleration, model year, origin_1, origin_2, origin_3) and 64 output neurons. If we multiple this, it is only 576. How did it become 640?


* Because there is one neuron added as *bias* in the input of each layer. So 9 + 1 = 10 multiplied by 64 is equal to 640.


* Same is true in the second layer which has 4,160 params. The input coming from the first layer is 64 + 1 bias = 65 multiplied by 64 = 4,160.


* The last layer has 64 neurons + 1 bias = 65 multiplied by 1 output neuron = 65.

```python
# You must install pydot (`pip install pydot`) 
# and install graphviz (see instructions at https://graphviz.gitlab.io/download/) , for plot_model/model_to_dot to work.
# conda install graphviz

from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

### Step 12. Train the model

* create a checkpoint for saving our model


* this way, we can save the best model throughout our training for future use


* create another callback to stop the training when it is not improving anymore


* patience is specify how many epochs to check if our monitor is improving


* we specify to monitor the validation loss

```python
checkpoint_name = "ModelMPG-{epoch:03d}--{val_loss:.5f}.hdf5"
# checkpoint = ModelCheckpoint(checkpoint_name, monitor="mean_absolute_error", verbose=1, save_best_only=True, mode="auto", save_freq="epoch")
checkpoint = ModelCheckpoint(checkpoint_name, monitor="mae", verbose=1, save_best_only=True, mode="auto")

early_stop = EarlyStopping(monitor="val_loss", patience=10)

#traing our model and save the history of the iterations
history = model.fit(
                    #features of our training data
                    #in this example, it would be the "size" and "room"
                    x_train,
                    #the label/target that we are trying to predict
                    # that would be "price" in this example
                    y_train,
    
                    callbacks = [checkpoint, early_stop],#save our model f
                    #a good practice is to create a validation data from the training data
                    #if we don't have a separate data for validation
                    #as noted earlier, we should not validate on the test data
                    #0.2 means that we will assign 10% of the training data as validation data
                    #you can set a higher percentage if we have more training data
                    #during the training, the validation data is NEVER used in the training
                    validation_split=0.2,
    
                    #the number of iterations over the entire training data
                    #this is hyperparameter and will check how to adjust it later
                    epochs=800,
    
                    #the number of samples in one epoch
                    #the higher the number the more accurate our model would be
                    #the higher the number the higher memory is used
                    #default size is 32
                    #since we have a very small traing dataset, we just process it as one batch
                    batch_size=250,
    
                    #verbosity of the training progress, set to 0 if you don't want to display anything
                    verbose=0)
```

### Step 13. Plot the result of our training history 

* plot the history of our training


* create a range of the number of epochs we used


* we can derive this from the length of history of our mean_absolute_error list


* take note that the list starts with an index 0 but we will start with 1 so, we add 1 to the length so as not to lose the last item in the list

```python
epochs = range(1, len(history.history["loss"]) + 1)

#let's define a figure with a wider width
plt.figure(figsize=(16,8))

#plot the training error
plt.plot(epochs, history.history["mae"], "b", label="Training MAE")
#together with validation error
plt.plot(epochs, history.history["val_mae"],"r", label="Validation MAE")

#write some labels
plt.title("History")
plt.xlabel = "Epochs"
plt.ylabel = "Error"

#display the legend
plt.legend()

#show the plot
plt.show()
```

By plotting the history, we can visualize the progress of our training. We can derive from the graph that at approximately at $610^{th}$ epoch, the error has flattened or converged.

### Step 14. Evaluate our model

```python
#evaluate the the data
y_predict = model.predict(x_test)

#compute and print the mean absolute error
loss = metrics.mean_absolute_error(y_test, y_predict)
print("Accuracy : +/- {0:.2f} MPG".format(loss))
```

6.	[2pts] The risk of vanishing gradients in deep networks. Select the spiral dataset (the bottom-right dataset under DATA), and change the network architecture to have four hidden layers with eight neurons each. Notice that training takes much longer and often gets stuck on plateaus for long periods of time. Also notice that the neurons in the highest layers (on the right) tend to evolve faster than the neurons in the lowest layers (on the left). This problem, called the vanishing gradients problem, can be alleviated with better weight initialization and other techniques, better optimizers (such as AdaGrad or Adam), or Batch Normalization.
